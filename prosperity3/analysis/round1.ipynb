```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

from prosperity3.analysis.data import read_price_data
```

```python
df = read_price_data([(1, -2), (1, -1), (1, 0)], lambda row: row.mid_price)
df
```

```python
prices = df.iloc[10000:20000]["SQUID_INK"]

rsi = 100 - (
    100
    / (
        1
        + prices.diff(1).mask(prices.diff(1) < 0, 0).ewm(alpha=1 / 14, adjust=False).mean()
        / prices.diff(1).mask(prices.diff(1) > 0, -0.0).abs().ewm(alpha=1 / 14, adjust=False).mean()
    )
)

fig = make_subplots(rows=2, cols=1)
fig.add_trace(go.Scatter(x=prices.index, y=prices, name="Mid price"), row=1, col=1)
fig.add_trace(
    go.Scatter(
        x=prices.index,
        y=((prices - prices.rolling(50).mean()) / prices.rolling(50).std()).rolling(100).mean(),
        name="Signal",
    ),
    row=2,
    col=1,
)
fig.update_layout(height=800)
fig.show()
```
