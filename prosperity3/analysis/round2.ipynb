```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

from prosperity3.analysis.data import read_price_data
```

```python
df = read_price_data([(2, -1), (2, 0), (2, 1)], lambda row: row.mid_price)
df
```

```python
data = df

basket1_diff = data["PICNIC_BASKET1"] - 6 * data["CROISSANTS"] - 3 * data["JAMS"] - data["DJEMBES"]
basket2_diff = data["PICNIC_BASKET2"] - 4 * data["CROISSANTS"] - 2 * data["JAMS"]

basket_diff = data["PICNIC_BASKET1"] - data["PICNIC_BASKET2"]
expected_basket_diff = 2 * data["CROISSANTS"] + data["JAMS"] + data["DJEMBES"]
basket_diff_delta = basket_diff - expected_basket_diff

basket1_jams = (data["PICNIC_BASKET1"] - 6 * data["CROISSANTS"] - data["DJEMBES"]) / 3
basket2_jams = (data["PICNIC_BASKET2"] - 4 * data["CROISSANTS"]) / 2
jams_diff = basket1_jams - basket2_jams

pd.DataFrame(
    {
        "basket1_diff": basket1_diff,
        "basket2_diff": basket2_diff,
        "basket_diff": basket_diff,
        "expected_basket_diff": expected_basket_diff,
        "basket_diff_delta": basket_diff_delta,
        "basket1_jams": basket1_jams,
        "basket2_jams": basket2_jams,
        "jams_diff": jams_diff,
    }
).describe()
```

```python
data = df.iloc[10000:20000]

basket1_diff = data["PICNIC_BASKET1"] - 6 * data["CROISSANTS"] - 3 * data["JAMS"] - data["DJEMBES"]
basket2_diff = data["PICNIC_BASKET2"] - 4 * data["CROISSANTS"] - 2 * data["JAMS"]

fig = make_subplots(rows=2, cols=1)
fig.add_trace(go.Scatter(x=data.index, y=data["PICNIC_BASKET1"], name="Price"), row=1, col=1)
fig.add_trace(go.Scatter(x=data.index, y=basket1_diff, name="Signal"), row=2, col=1)
fig.update_layout(height=800)
fig.show()
```
