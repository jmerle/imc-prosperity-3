```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
reserve_prices = np.random.uniform(low=160, high=160 + 110, size=1000_000)
reserve_prices[reserve_prices > 200] += 50
px.histogram(reserve_prices[:100_000]).show()
```

```python
lows = []
highs = []
profits = []
ppt_takens = []
ppt_totals = []

take_bid = [bid > reserve_prices for bid in range(160, 321)]
total_turtles = len(reserve_prices)

avg_high = 288

for low in range(160, 321):
    take_low = take_bid[low - 160]
    low_turtles_taken = take_low.sum()
    low_profit = low_turtles_taken * (320 - low)

    for high in range(low, 321):
        high_turtles_taken = (take_low ^ take_bid[high - 160]).sum()

        if high >= avg_high:
            p = 1
        else:
            p = ((320 - avg_high) / (320 - high)) ** 3

        profit = low_profit + high_turtles_taken * (320 - high) * p
        ppt_taken = (
            profit / (low_turtles_taken + high_turtles_taken) if (low_turtles_taken + high_turtles_taken) > 0 else 0
        )
        ppt_total = profit / total_turtles

        lows.append(low)
        highs.append(high)
        profits.append(profit)
        ppt_takens.append(ppt_taken)
        ppt_totals.append(ppt_total)

df = pd.DataFrame({"low": lows, "high": highs, "profit": profits, "ppt_taken": ppt_takens, "ppt_total": ppt_totals})
df.sort_values("profit", ascending=False)
```
